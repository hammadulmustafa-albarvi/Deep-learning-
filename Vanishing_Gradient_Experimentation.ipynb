```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
```

```python
X,y = make_moons(n_samples=250, noise=0.05, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1])
```

# Test with alot of hidden layers

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
model = Sequential()
```

```python
model.add(Dense(10,activation='sigmoid',input_dim=2))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
old_w = model.get_weights()[0]
```

```python
# model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=50)
```

```python
new_w = model.get_weights()[0]
```

```python
change_in_percentage = ((new_w-old_w)/old_w )* 100
```

```python
new_w
```

```python
change_in_percentage
```

# Test with few Hidden Layers

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
model = Sequential()
model.add(Dense(10,activation='sigmoid',input_dim=2))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
old_w = model.get_weights()[0]
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=50)
new_w = model.get_weights()[0]
change_in_percentage = ((new_w-old_w)/old_w )* 100
```

# Test with changed Activation Function

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
model = Sequential()
model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

old_w = model.get_weights()[0]
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=50)
new_w = model.get_weights()[0]
change_in_percentage = ((new_w-old_w)/old_w )* 100
```
